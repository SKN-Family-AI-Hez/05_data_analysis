## DataFrame 생성

```python
# dict -> df 생성
data = {"one": [1.0, 2.0, 3.0], "two": [10, 20, 30], "three": 99}

# dict 의 각 item이 하나의 series로 처리됨
df = pd.DataFrame(data) # 차수가 다르면 자동으로 값이 채워지는 브로드캐스팅 발생
print(df['one']) # 'one'열의 Series 선택
df
```

```python
# 여러 개의 dict -> dataframe
data : list[dict[str,int]] = [{
    "a": 10,
    "b": 20,
}, {
    "a": 30,
    "b": 40,
    "c": 50,
}]

# 하나의 dict -> 하나의 레코드, key는 컬럼으로 처리
df = pd.DataFrame(data)
df # 0 번행의 C 열의 NaN == 결측치
```

```python
# ndarray -> df
# 컬럼이름, 인덱스 이름 라벨링
arr = np.arange(9).reshape(3,3)
df = pd.DataFrame(arr, columns=('a','b','c'), index = ['1일','2일','3일'])
df
```

```python
# ser -> df 시리즈로 데이터프레임 만들기
#시리즈를 두개 만든 다음에 키 값을 넣어서 딕셔너리 만들기
name_ser = pd.Series(["홍길동", "이순신", "강감찬"])
age_ser = pd.Series([20, 25, 30])

df = pd.DataFrame({"name": name_ser, "age": age_ser})
df
```

```python
# tuple(record) -> dataframe
# 하나의 객체

data = [
    ('홍길동', 20),
    ('이순신', 30),
    ('논개', 43),
]
df = pd.DataFrame.from_records(data, columns=['이름', '나이'])
df
```

## 조회
- 행열조회
    - df.iloc[인덱스]
    - df.loc[라벨]
- 열조회
    - df[컬럼명]

```python
dicts = [
    {'name' : '홍길동', 'job' : '강사', 'age' : 33},
    {'name' : '신사임당', 'age' : 44, 'job' : '영업사원'},
    {'name' : '이순신', 'age' : 38, 'job' : '장군'},
    {'name' : '논개', 'age' : 25, 'job' : '개발'},
    {'name' : '강감찬', 'age' : 47, 'job' : '교수'},
]
df = pd.DataFrame(dicts)
df
```

### iloc (인덱스 위치)

```python
# 홍길동 행 조회
print(df.iloc[0])
print()
print(type(df.iloc[0])) # Series, df에서 한 행 또는 한 열은 Series 타입니다.
print()
print(df.iloc[0].index)
print()
print(df.iloc[0].name) # df의 인덱스 == ser 의 이름
```

```python
# 슬라이싱 지원
df2 = df.iloc[0:2] # 0,1 행 반환
df2
```

```python
# fancy indexing    지원
df.iloc[[0,2]]
```

```python
# 단일 값 조회
print(df.iloc[0,1])

# 여러 열 조회
df.iloc[:2, 1]
```

```python
df.iloc[:2, :2] # 0-1번행, 0-1범열
print(type(df2))
df2
```

### loc

```python
# 인덱스에 대입, 인덱스명 지정 == 라벨링
df.index = ['client1', 'client2', 'client3', 'client4', 'client5']
df
```

## loc 슬라이싱은 end 값을 포함한다

```python
# loc 를 이용한 컬럼 단일 값 조회
print(df.loc['client2', 'job'])

# loc를 이용한 슬라이싱
print(df.loc['client2':'client4'])
print()

# loc를 이용한 슬라이싱 + 컬럼 선택 - client 2,3,4,행의 job만 보기
print(df.loc['client2':'client4', 'job'])
print()


# loc를 이용한 슬라이싱 + 컬럼 선택 - client 2,3,4,행의 job + age보기
print(df.loc['client2':'client4', 'job':])
print()

```

### 컬럼라벨

```python
# df['컬럼명'] : 해당 컬럼 시리즈 반환
print(type(df['name'])) # 가로든 세로든 한 줄이면 시리즈임.
df['name']
```

```python
# 컬러라벨 fancy indexing
print(type(df[['name', 'job']] ))
df[['name', 'job']] # 한줄이 아니다 -> 시리즈의 묶음 : dataframe 타입
```

```python
# loc + 컬럼라벨( boolean indexing)
print(df['name'] == '이순신') # 불리언인덱싱
print()
print(df.loc[df['name'] == '이순신']) # 팬시 인덱싱
```

```python
# boolean indexing (비교연산)
print(df['age'] > 30)
df[df['age'] > 30]
```

```python
# 30세 미만, 개발 직종을 갖는 행 조회
df.loc[ ( df['age'] < 30 ) & ( df['job'] == '개발' ) ]
```

```python
# - 이름에 '신'이 들어가거나 나이가 40대이상 조회
df[(df['age'] >= 40) | (df['name'].str.contains('신'))]
```

## 행 추가/삭제

```python
students = [
    {'name': '홍길동', 'midterm': 95, 'final': 85},
    {'name': '신사임당', 'midterm': 85, 'final': 80},
    {'name': '이순신', 'midterm': 10, 'final': 30}
]
df = pd.DataFrame(students)
df
```

```python
# 중복되니 않는 인덱스에 한 행(Series타입) 추가(대입)
next_index = len(df)

df.loc[next_index] = pd.Series({'name': '강감찬', 'midterm': 88})  # 추가된 행에 final컬럼값 누락 -> NaN (결측치)
df
```

```python
# 중복되는 인덱스에 값 추가 = 대체됨
next_index = 3

df.loc[next_index] = pd.Series({'name': '율곡이이', 'midterm': 77})  # 추가된 행에 final컬럼값 누락 -> NaN (결측치)
df
```

```python
# 행삭제 -> 원본 유지
# display( df.drop(3)
# display(df)
df = df.drop(3)
df
```

```python
df.loc[:, 'phone'] = ['010-1234-5678', ['010-3333-4444']]
df
```

```python
# 특정 조건을 만족하는 모든 행 삭제
df = df[df['final'] >= 60 ] # 60점 이상만 남기기 == 60점 미만 삭제
df
```

## 컬럼 추가/삭제

```python
# 중간고사, 기말고사 평균 컬럼(mean)을 추가
df.loc[:, 'mean'] = (df['midterm'] + df['final']) / 2
df
```

```python
# 컬럼삭제
df[['name', 'midterm', 'final']] # fancy indexing
```

```python
def pass_or_fail(value):
    return 'P' if value >= 60 else ('F')

df.loc[:, 'avg'] = (df['midterm'] + df['final']) / 2

df.loc[:, 'pass_fail'] = df['avg'].apply(pass_or_fail)
df
```

```python
# apply 함수에 행 (record) 전체를 인자로 전달해서 처리하기
def pass_or_fail2(record):
    avg = (record['midterm'] + record['final']) / 2
    return 'P' if avg >= 60 else 'F'

df.loc[:, 'pass_fail2'] = df.apply(pass_or_fail2, axis = 1)
```
